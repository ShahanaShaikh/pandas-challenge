```python
import pandas as pd 
```

```python
purchase_file = "HeroesOfPymoli.csv"
```

```python
py_df = pd.read_csv(purchase_file)
py_df.head()
```

```python
# Total Number of Players
py_df["SN"].count()
py_df.drop_duplicates()
total_players = py_df["SN"].count()
print(total_players)
```

```python
# Number of Unique Items
unique_items = len(py_df["Item ID"].unique())

# Average Purchase Price
avg_pp = py_df["Price"].mean()

# Total Number of Purchases
total_purchases = py_df["Price"].count()

# Total Revenue
total_rev = py_df["Price"].sum()

summary_table = pd.DataFrame({"Number of Unique Items": unique_items,
                              "Average Purchase Price": [avg_pp],
                              "Total Number of Purchases": [total_purchases],
                              "Total Revenue": [total_rev]})

summary_table
```

```python
# Percentage and Count of Male Players
# Percentage and Count of Female Players
# Percentage and Count of Other / Non-Disclosed

gender_dg = py_df["Gender"].value_counts()

player_percentage = gender_df/total_players

gender_breakdown = pd.DataFrame({"Count by Gender": gender_dg, "Percentage of Players": player_percentage})
gender_breakdown["Percentage of Players"] = gender_breakdown["Percentage of Players"].map("{:,.2%}".format)

gender_breakdown
```

```python
#The below each broken by gender

# Purchase Count
# Average Purchase Price
# Total Purchase Value
# Average Purchase Total per Person by Gender

female_purchase_count = len(py_df.loc[py_df["Gender"]== "Female",:])
male_purchase_count = len(py_df.loc[py_df["Gender"]== "Male",:])
other_purchase_count = len(py_df.loc[py_df["Gender"]== "Other",:])


female_avg_pp = py_df.loc[py_df["Gender"]== "Female",["Price"]].mean()
male_avg_pp = py_df.loc[py_df["Gender"]== "Male",["Price"]].mean()
other_avg_pp = py_df.loc[py_df["Gender"]== "Other",["Price"]].mean()

female_total_pp = py_df.loc[py_df["Gender"]== "Female",["Price"]].sum()
male_total_pp = py_df.loc[py_df["Gender"]== "Male",["Price"]].sum()
other_total_pp = py_df.loc[py_df["Gender"]== "Other",["Price"]].sum()

female_avg_purchase_total = female_total_pp/female_purchase_count
male_avg_purchase_total = male_total_pp/male_purchase_count
other_avg_purchase_total = other_total_pp/other_purchase_count

gender_analysis = pd.DataFrame([
    {"gender":"Female","purchase_count": female_purchase_count,
    "average_purchase_price": "${:,.2f}".format(female_avg_pp[0]),
    "total_purchase_value": "${:,.2f}".format(female_total_pp[0]),
    "avg_purchase_total": female_avg_purchase_total},
     {"gender":"Male","purchase_count": male_purchase_count,
    "average_purchase_price": "${:,.2f}".format(male_avg_pp[0]),
    "total_purchase_value": "${:,.2f}".format(male_total_pp[0]),
    "avg_purchase_total": male_avg_purchase_total},
    {"gender":"Other","purchase_count": other_purchase_count,
    "average_purchase_price": "${:,.2f}".format(other_avg_pp[0]),
    "total_purchase_value": "${:,.2f}".format(other_total_pp[0]),
    "avg_purchase_total": other_avg_purchase_total}
    ])

gender_analysis
```

```python
# The below each broken into bins of 4 years (i.e. <10, 10-14, 15-19, etc.)

age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

py_df["Ranges"] = pd.cut(py_df["Age"], age_bins, labels = group_names)
demographic_group = py_df.groupby("Ranges")

py_df.head()
```

```python
# Purchase Count
# Average Purchase Price
# Total Purchase Value
# Average Purchase Total per Person by Age Goup

age_totals = py_df["Ranges"].value_counts()
age_percentage = age_totals / total_players
age_analysis = pd.DataFrame({"Total Count": age_totals, "Percentage of Players": age_percentage})

age_analysis['Percentage of Players'] = age_analysis['Percentage of Players'].map("{:,.2%}".format)


age_analysis = age_analysis.sort_index()
age_analysis


#age_purchase_count = py_df.groupby(["Ranges"]).count()["Price"]
#age_avg_purchase_price = py_df.groupby(["Ranges"]).mean()["Price"]
#age_total_purchase_value = py_df.groupby(["Ranges"]).sum()["Price"]
#age_avg_purchase_total = age_total_purchase_value/py_df.groupby(["Ranges"]).count()

#age_analysis = pd.DataFrame({"Purchase Count": age_purchase_count, 
                             # "Average Purchase Price": "${:,.2f}".format(age_avg_purchase_price[0]), 
                             # "Total Purchase Value": "${:,.2f}".format(age_total_purchase_value[0]), 
                             # "Categorized Avg Purchase Total": age_avg_purchase_total})

# age_analysis
```

```python
py_df["Ranges"] = pd.cut(py_df["Age"], age_bins, labels=group_names)


age_purchase_count = py_df.groupby(["Ranges"]).count()["Price"]
age_avg_purchase_price = py_df.groupby(["Ranges"]).mean()["Price"]
age_total_purchase_value = py_df.groupby(["Ranges"]).sum()["Price"]
avg_purchase_total_gender_age = age_total_purchase_value/age_analysis["Total Count"]


age_data_analysis = pd.DataFrame({"Purchase Count": age_purchase_count, 
                                  "Average Purchase Price": age_avg_purchase_price, 
                                  "Total Purchase Value": age_total_purchase_value, 
                                  "Avg Purchase Total per Person by Age": avg_purchase_total_gender_age})


age_data_analysis["Average Purchase Price"] = age_data_analysis["Average Purchase Price"].map("${:,.2f}".format)
age_data_analysis["Total Purchase Value"] = age_data_analysis["Total Purchase Value"].map("${:,.2f}".format)
age_data_analysis ["Purchase Count"] = age_data_analysis["Purchase Count"].map("{:,}".format)
age_data_analysis["Avg Total Purchase per Person"] = age_data_analysis["Avg Purchase Total per Person by Age"].map("${:,.2f}".format)
age_data_analysis 
```
